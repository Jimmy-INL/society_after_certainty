## Certainty

Certainty holds a very provisionally important place in scietific research and mathematical modeling. When we pursue a line of inquiry, we must first stabilize a system of knowledge and hold constant the assumptions and tools we want to not "move" while we're examining a phenomona in motion. We will never be able to fix reality without losing vital information about what came before and what came after. The context is important, so we understand all observable information to be a part of a moving context, with tools/lenses/assumptions firmed up a priori. This is includes the assumptions baked into the traditions of statistical analysis we chose to use. And yes, you have a choice! Statistics, like it's big sibling mathematics, is filled with orientation choices. So that's the positive application of certainty in research - sealing up our vessel to brave uncharted waters. 

As we progress through this book I'd like you to keep tabs with me about all the assumptions we must make along the way as we journey away from life's complexity, to the rough-and-tumble world of modeling. I invite you to do this with me so you can build your own practice of critiquing this enterprise we call modeling. It can be tempting to believe you have found "truth" with your model, but does an accurate sonic printout of a mother singing a lullaby to her child capture all of the truth of that complex exchange? The answer is no, but rather than know the boundaries of certainty as a mandate you must uphold, I encourage you to find your own practice.

## Uncertainty

Let us now consider examine two ways of thinking of uncertainty:

### Aleotoric

This is statistical uncertainty in our research that we can't pin down. A noise we accept. 

### Epistemic

Systematic uncertainty is uncertainty that arises from our process, even if we, in theory, have some certainty about the concepts themselves. These are the pesky lurking variables/phenomena that get in the way of our tidy theories we've already proved. Contextual things that in themselves represent new information.  

## Frequentist Tools

### Central Limit Theorem

So what can we say about samples that we draw from populations of data?

Well, a pillar of probability theory, the Central Limit Theorem, may shed some light. This states, "given certain conditions, the arithmetic mean of a sufficiently large number of iterates of independent random variables, each with a well-defined (finite) expected value and finite variance, will be approximately normally distributed, regardless of the underlying distribution" [Wikipedia] "As n gets larger, the distribution of the difference between the sample average Sn and its limit µ, when multiplied by the factor √n (that is √n(Sn − µ)), approximates the normal distribution with mean 0 and variance σ2"

"Whatever the form of the population distribution, the sampling distribution tends to a Gaussian, and its dispersion is given by the Central Limit Theorem"
